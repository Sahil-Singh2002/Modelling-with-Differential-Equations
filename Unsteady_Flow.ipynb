```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Given constants
u0 = 2.0
kt = 1.0

# Function defining the velocity components u and v
def velocity_components(x, y):
    u = u0 * x
    v = kt * y
    return u, v

# Parameters for plotting
x_range = np.linspace(-4, 4, 400)
y_range = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x_range, y_range)

# Calculate velocity components at each grid point
U, V = velocity_components(X, Y)
```

```python
# Part (a): Streamline equation
fig, ax = plt.subplots(figsize=(8, 8))
ax.streamplot(X, Y, U, V, density=1.5, linewidth=1, color='b')

ax.set_title('Streamlines')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```

```python
# Part (b): Sketch streamlines for different times
times = [0, 1, 2]
x_range = np.linspace(-4, 4, 100)
y_range = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x_range, y_range)



fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for t, ax in zip([0, 1, 2], axes):
    x_t = X - u0 * t
    y_t = Y - kt * t
    U_t, V_t = velocity_components(x_t, y_t)
    ax.streamplot(X, Y, U_t, V_t, density=1.5, linewidth=1, color='b')
    ax.set_title(f'Streamlines at t = {t}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
plt.tight_layout()
plt.show()
```

```python
# (c) Pathlines (parabolic paths)

fig, ax = plt.subplots(figsize=(8, 8))
x0 = 1.0
y0 = 0.5
t_values = np.linspace(0, 4, 100)
x_path = x0 + u0 * t_values
y_path = y0 + kt * t_values
ax.plot(x_path, y_path, color='r')
ax.set_title('Pathline for t >= 0')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
plt.grid()
plt.show()
```

```python
# (d) Streakline at t = 1
fig, ax = plt.subplots(figsize=(8, 8))
x0 = 1.0
y0 = 0.5
x_streak = X - u0 * 1
y_streak = Y - kt * 1
U_streak, V_streak = velocity_components(x_streak, y_streak)
ax.streamplot(X, Y, U_streak, V_streak, density=1.5, linewidth=1, color='g')
ax.plot(x_path, y_path, color='r', label='Pathline')
ax.plot(x0, y0, marker='o', color='k', label='Release Point')
ax.set_title('Streakline at t = 1')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.legend()
plt.grid()
plt.show()
```
